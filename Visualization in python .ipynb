# Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
status=['Divorced','single','married','widowed']
y_pos=np.arange(len(status))
number=[400,556,678,123]
plt.bar(y_pos,number,align='center',alpha=0.5)
plt.xticks(y_pos,status)
plt.ylabel("count")
plt.title("Martial status.......")
plt.show()
```

```python
MySubject=['Python','R language','Machine Learning','SQL']
y_pos_s=np.arange(len(MySubject))
number_s=[80,90,95,100]
plt.bar(y_pos_s,number_s,align='center',alpha=1,facecolor='red')
plt.xticks(y_pos_s,MySubject)
plt.ylabel("count")
plt.title("My Subject All")
plt.show()
```

```python
pokeman['Color'].value_counts().plot.bar()
plt.ylabel("count")
plt.title("Pokeman dataset color")
plt.show()
```

```python
pokeman['Color'].value_counts().nlargest(5).sort_values(ascending=True).plot.barh()
plt.title("Pokeman dataset all")
plt.show()
```

# Pie Chart

```python
labels='zeeshan','raja','elai','buner'
sizes=[215,434,300,455]
explode=(0,0.1,0,0)
plt.pie(sizes,explode=explode,labels=labels, shadow=True, startangle=140)
plt.axis("equal")
plt.show()
```
